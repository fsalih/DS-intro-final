```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score

from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
```

```python
RS = 42
```

```python
sessions = pd.read_csv('test_out.csv', index_col=0)
```

```python
sessions
```

```python
# V.2
columnst_to_transform = [ 'utm_source', 'utm_medium', 'utm_campaign', 'utm_adcontent',
       'utm_keyword', 'device_category', 'device_os', 'device_brand',
       'device_screen_resolution', 'device_browser', 'geo_country', 'geo_city']
```

```python
ohe = OneHotEncoder(sparse=False)
```

```python
ohe.fit(sessions[columnst_to_transform])
```

```python
sessions[ohe.get_feature_names_out()] = ohe.transform(sessions[columnst_to_transform])
```

```python
sessions = sessions.drop(columns=columnst_to_transform)
```

```python
X = sessions.drop('target', axis=1)
y = sessions['target']
```

```python
%%time
x_tr, x_test, y_tr, y_test = train_test_split(sessions.drop(columns='target'), sessions['target'], random_state=RS)
```

```python
logreg = LogisticRegression(class_weight='balanced',
                           random_state=RS)
```

```python
%%time
logreg.fit(x_tr, y_tr)
```

```python
y_pred_lr = logreg.predict(x_test)
roc_auc_score(y_test, y_pred_lr)
```

```python
%%time
logreg.fit(X, y)
```

```python
y_pred_lr = logreg.predict(X)
roc_auc_score(y, y_pred_lr)
```
